```python
#Imports Graphical Features Extension to be called as "plt"
import matplotlib.pyplot as plt
 
#Initializes x Value (x-values for charts) as 0
x = 0

#Gathers user input for lower end of the domain of trial cases
print("Enter the lower bounds of the interval")
counter = int(input())
 
#Gathers user input for upper end of the domain of trial cases
print("Enter the upper bounds of the interval")
repetition_count = int(input())
 
#Initialize Trial Counter
origin_counter = counter

#Initializes the Highest Num (Greatest Y-Value) as the initial input the user provided for counter (Lower bounds of the Domain)
highest_num = counter

#Initializes the great interval (Trial that produces the greatest Y-Value recorded) to 0 
great_interval = 0

#Initialize (x,y) for first Trial Case
yval = [counter]
xval = [x]
 
while origin_counter != repetition_count + 1:
 
    x = 0
    print("\n\n")
    print("Trial Number: ", origin_counter )
    
    while counter != 1:
        
        if counter % 2 == 0:
            print(1,end="")
            counter = counter // 2
            
        else:
            print(0)
            counter = (counter * 3) + 1
 
        yval.append(counter)
        x += 1
        xval.append(x)
        plt.plot(xval, yval)
        if (highest_num < counter):
            highest_num = counter
            great_interval = origin_counter
        
    
    yval.clear()
    xval.clear()
    origin_counter += 1
    counter = origin_counter

    
plt.grid(color='black', linestyle='-', linewidth=0.2)
plt.show()
print("Greatest Value: ", "{:,}".format(highest_num))
print("Initial Start Value: ", great_interval)
```
